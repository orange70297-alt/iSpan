```python
# 建立 function 3
# 給參數
def greeting(msg, name):
    str_list = msg.split()
    str_list.append(name)
    return str_list

data = greeting("Hello World!", "Jarvis")
print(data)
data = greeting(name="Jarvis", msg="Hello World!")
print(data)
```

```python
# 建立 function 4
# 預設參數值
def greeting(name: str, msg="Good Morning") -> list[str]:
    str_list = msg.split()
    str_list.append(name)
    return str_list

data = greeting(name="User")
print(data)
data = greeting("User", "Hello, World!")
print(data)
```

```python
# 改寫 Module2 practice
def get_discount(total, discount):
    if total >= 2000:
        total -= discount * (total//2000)

    return total

data = [
    {
        "total": 6000,
        "discount": 200
    },
    {
        "total": 8000,
        "discount": 200
    }
]

for d in data:
    # print(d["total"], d["discount"])
    answer = get_discount(d["total"], d["discount"])
    print(f"折扣後金額: {answer}元")
```
