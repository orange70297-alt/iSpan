```python
print(a)
print("程式執行完畢")
```

```python
# 可以連續定義數個例外處理機制
# 最符合的例外事件類別會先拋出
try:
    num_list = [1, 2]
    print(num_list[2])
    print(a)
    x = 10
    print(x)
except NameError:
    print("發生 NameError 錯誤")

print("程式執行完畢")
```

```python
try:
    num_list = [1, 2]
    print(num_list[2]) # 發生錯誤 進入 except
    print(a) # 不會執行的區塊
    x = 10   # 不會執行的區塊
    print(x) # 不會執行的區塊
except Exception as e:
    print(f"發生 {e} 錯誤")

print("程式執行完畢")
```
