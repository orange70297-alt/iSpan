百貨公司周年慶 消費金額超過 2000 要折 200  

請撰寫一支程式  
計算客戶消費的結帳金額

```python
total = 5500
if total >= 2000:
    total -= 200 # total = total - 200
print(total)
```

滿 4000 折 400 滿 6000 折 600.....

```python
total = 6500
discount = 200
if total >= 2000:
    # cnt = total // 2000
    # total -= discount * cnt

    total -= discount * (total//2000)

print(total)
```
