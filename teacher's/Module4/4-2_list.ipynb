```python
# 刪除索引位置的元素
# list.pop(<index>)
ids = [1, 2, 3, 4, 5, 6]
ids.pop(3)
print(ids)
```

```python
# 在指定索引 1, 新增元素 9 ，其餘元素往後移
# list.insert(<index>, <value>)
ids = [1, 2, 3, 4, 5, 6]
ids.insert(1, 9)
print(ids)
```

```python
# 移除指定的值
# list.remove(<value>)
ids = [1, 2, 3, 4, 5, 6]
ids.remove(4)
print(ids)
```

```python
ids = [1, 2, 3, 4, 5, 6]
for _ in range(5):
    ids.insert(1, 9)
print(ids)
```

```python
while 9 in ids:
    ids.remove(9)

print(ids)
```

```python
ids = [1, 2, 3, 4, 5, 6]
for i in range(5):
    ids.append(10)
print(ids)

# 計算 list 內的某個值 共有幾個
# list.count(<value>)
cnt_10 = ids.count(10)
print(cnt_10)
```

### list 搭配 for 迴圈

```python
# 建立一個 1~10 的 list
num_list = []
for i in range(1, 11):
    num_list.append(i)
print(num_list)
```

```python
# 建立一個 1~50 的 list 只要偶數
num_list = []
for i in range(1, 51):
    if i % 2 == 0:
        num_list.append(i)
print(num_list)
```

```python
outside = []
inside = []
for i in range(1, 11):
    inside.append(i)
print(inside, id(inside))
```

```python
outside = []
inside = []
for i in range(1, 11):
    inside.append(i)
    outside.append(inside.copy())

print(outside)
```

```python
x = [i for i in range(10)]
num = x.append(11)
print(num)
```

```python
# 建立一個 list 裡面要放遞增的 list
# 例如 [[1], [1, 2], [1, 2, 3]...[1, 2,... 10]]
outside = []


for i in range(1, 11):
    inside = []
    # print(f"i: {i}")
    for c in range(1, i+1):
        # print(f"c: {c}")
        inside.append(c)
    print(inside, id(inside)) # id() 取得變數的記憶體位置
    outside.append(inside)


print(outside)
```

```python
print(outside[0], id(outside[0]))
```

```python
# enumerate() 同時取得 index 與 value
ids = [10, 2, 35, 4, 15]
for index, value in enumerate(ids):
    print(f"index: {index}, value: {value}")
```

### List comprehension 列表推導式
`new_list = [expression for item in iterable [if condition]]`  
  
`list的名稱 = [表達式 for 元素 in 迭代對象 [if 判斷式(...的條件)]]`

```python
# 建立一個 1~10 的 list
num_list = []
for i in range(1, 11):
    num_list.append(i)
print(num_list)
```

```python
# list comprehension
num_list_1_to_10 = [i for i in range(1, 11)]
print(num_list_1_to_10)
```

```python
# 1~10 的二次方的 list
data = [i**2 for i in range(1, 11)]
print(data)
```

```python
outside = []
for i in range(1, 11):
    inside = []
    for c in range(1, i+1):
        inside.append(c)
    outside.append(inside)
print(outside)
```

```python
outside = []
for i in range(1, 11):
    outside.append([c for c in range(1, i+1)])
print(outside)
```

```python
outside = [[c for c in range(1, i+1)] for i in range(1, 11)]
print(outside)
```

```python
# 建立一個 1~50 的 list 只要偶數
num_list = []
for i in range(1, 51):
    if i % 2 == 0:
        num_list.append(i)
print(num_list)
```

```python
num_list = [i for i in range(1, 51) if i % 2 == 0]
print(num_list)
```
