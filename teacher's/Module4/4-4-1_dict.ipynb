```python
# 建立一個 value 含有 dict 的 dict
user_info = {
    "name": "Jarvis",
    "age": 20,
    "info": {
        "nickname": "Lu",
        "favorite_role": "ironman",
        "phone": [
            "0912345678",
            "0912345679"
        ]
    }
}
```

```python
# 取得 nickname
print(user_info["info"])
print(user_info["info"]["nickname"])
```

```python
# 取得所有 phone_number
print(user_info["info"]["phone"])
for phone_number in user_info["info"]["phone"]:
    print(phone_number)
```

```python
'''
改寫 fruit_dict 每個水果需要 price, quantity/amount, color
apple 有 綠色 和 紅色
orange 有 橘色
tomato 有 紅色
'''
fruit_dict = {
    "apple": {
        "price": 30,
        "amount": 100,
        "color": ["green", "red"]
    },
    "orange": {
        "price": 50,
        "amount": 30,
        "color": ["orange"]
    },
    "tomato": {
        "price": 100,
        "amount": 200,
        "color": ["red"]
    }
}


print(fruit_dict)
```

##### dict 的合併

```python
# dict 的合併
# dict.update()
fruit_dict2 = {
    "kiwi": {
        "price": 30,
        "amount": 20,
        "color": ["green"]
    }
}

fruit_dict.update(fruit_dict2)
print(fruit_dict)
```

```python
print(fruit_dict)
fruit_dict3 = {
    **fruit_dict, # 補充 看過有印象就好
    "kiwi": {
        "price": 100,
        "amount": 50,
        "color": ["yellow"]
    },
}
print(fruit_dict3)
```

##### 另外的 print 方式 `pprint` (pretty-print)
