|string方法| 功能 |
|:---:|:---:|
|`str.replace()` | 替換字串 |
|`str.strip()` | 去除兩側空格 |
|`str.upper()` | 字串小寫變大寫 |
|`str.lower()` | 字串大寫變小寫 |

```python
# 字串變數初始化
name = 'Jarvis'
print(name)
print(name[0])

name = "Larvis"
print(name)
```

```python
# 替換字串
# 方法 也是 function 的一種
# 定義在 class 底下的 function 可以稱為 method
'''
str.replace()
將 string 中的 str1 替換成 str2
'''
name = 'Alex ex'

new_name = name.replace('ex', 'len')
print(new_name)
print(name)
```

```python
# 去除兩側空格
'''
str.strip()
去除字串 string 左、右兩邊的空格
'''
data = "  |__ data __|  "
print(data)

new_data = data.strip()
print(new_data)

new_data = new_data.replace(' ', '')
print(new_data)

new_data = new_data.replace('|', '')
print(new_data)

new_data = new_data.replace('_', '')
print(new_data)
```

```python
data = "1  |__ data __|  1"
print(data)

new_data = data.rstrip('1') # 去除右邊
# new_data = data.lstrip('1') # 去除左邊
print(new_data)
```

```python
# 字串小寫變大寫
'''
str.lower()
將字串 string 裡的字母全部改成小寫
'''
data = "CAR"
new_data = data.lower()
print(new_data)
'''
str.upper()
將字串 string 裡的字母全部改成大寫
'''
print("good".upper())
```

```python
# 字串分割成 list
# str.split()
greeting = "hello world! hello John"
str_list = greeting.split('!')
print(str_list)
```

```python
for data_str in str_list:
    print(data_str)
    for char in data_str:
        print(char, end=', ')
    print()
```
