### `json` 套件

| 用法 | 說明 |
|:--:|:--:|
|`json.dump(<obj>, <file>, [indent=<num>])` |物件轉JSON並寫入文件|
|`json.dumps(<obj>)` |物件轉JSON字串|
|`json.load(<file>)` |讀取JSON文件|
|`json.loads(<str>)` |將JSON字串轉成Python物件|

```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange']
    }
}
print(fruit_dict)
```

```python
# 引入套件
import json
# json.dumps 物件轉 JSON 字串
with open('fruit_data.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(fruit_dict))
```

```python
# json.dump() 物件轉 JSON 之後直接寫入檔案
with open('fruit_data.json', 'w', encoding='utf-8') as f:
    json.dump(fruit_dict, f, indent=4)
```

```python
# json.loads() 讀取字串轉物件(list/dict)
with open('fruit_data.json', 'r', encoding='utf-8') as f:
    data = f.read()
    data_dict = json.loads(data)

print(data_dict)
print(data_dict["apple"])
```

```python
x = "[1, 2, 3]"
print(type(x))

x_list = json.loads(x)
print(type(x_list))
print(x_list[0])
```

```python
# json.load() 讀取檔案直接轉物件
with open('fruit_data.json', 'r', encoding='utf-8') as f:
    data_dict = json.load(f)

print(data_dict)
print(data_dict["apple"])
```
