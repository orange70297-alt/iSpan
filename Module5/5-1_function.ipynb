```python
# 建立 function 3
def greeting(msg, name): #不只一個參數用逗號隔開
    str_list = msg.split()
    # print(str_list)
    str_list.append(name) 
    return str_list
data = greeting("hello hello world","Jarvis")
print(data)
data = greeting(name= "Jarvis", msg="hey Hello World")

print(data)
```

```python
# 建立 function 4
# 預設參數值

def greeting(name, msg="Good Morning"): # 有預設參數的要放後面
    str_list = msg.split()
    print(str_list)
    str_list.append(name)
    return str_list

data = greeting("User")
print(data)
data = greeting("user","Hello,World")
print(data)
data = greeting("Tim","ni hao")
print(data)
```

```python
def get_discount(total,discount):
    if total>=2000:
        total -= discount* (total//2000)
    return total

data = [
    {
        "total": 6000 ,
        "discount": 200
     },

    {   
        "total": 8000,
        "discount": 200
    }
]

for d in data:
    print(d["total"],d["discount"])   # ["total"] ["discount"] -> 取出字典'key'的value值
    answer = get_discount(d["total"],d["discount"])
    print(f"折扣後的金額: {answer} 元")
```
