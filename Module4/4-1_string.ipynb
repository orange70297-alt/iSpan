|string方法| 功能 |
|:---:|:---:|
|`str.replace()` | 替換字串 |
|`str.strip()`| 去除兩側空格 |
| `str.upper`| 字串小寫變大寫 |
| `str.lower`| 字串大寫變小寫 |

```python
# 字串變數初始化
name = 'Jarvis'
print(name)
print(name[1])

```

```python
# 替換字串
# 方法 也是 function 的一種
# 定義在 class 底下的 funtion 可以稱為 method
'''
str.replace()

將 string 中的 str1 替換成 str2
'''
name = 'Alex ex'
new_name = name.replace("ex","len")

print(name)
print(new_name) 
```

```python
# 去除兩側空格
'''
str.strip()
去除字串 string 左、右兩邊的空格
'''
data = ' |__data __|  '
print(data)
new_data = data.strip()   #strip (去除兩邊空格)
print(new_data)

new_data = new_data.replace(' ','')  
print(new_data)
new_data = new_data.replace('|','')
print(new_data)
new_data = new_data.replace('_','')
print(new_data)
```

```python
# 字串小寫變大寫
'''
str.lower()

將字串 string 裡的字母全部改成小寫
'''
data = "CAR"
new_data = data.lower()
print(new_data)
'''
str.upper()
將字串 string 
'''
new_data1=new_data.upper()
print(new_data1)
print(new_data)

print("good".upper()) # 會重新return值 才能這樣打印 (list不行)
```

```python
# 字串分割成 list
# str.split()
greeting = "hello world! hello John"
str_list = greeting.split("!")
print(str_list)
```
