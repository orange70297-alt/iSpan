```python
# 初始化一個 list
# 因為list可變,所以不前面不需要放變數接附值
# 它可以一直改變list內的資料 

ids = [1, 2, 3, 4, 5, 6]

print(ids)
```

```python
# 新增元素在 list 尾端
# 非常常用
# list,append()
ids = [1, 2, 3, 4, 5, 6]
ids.append(7)
print(ids)
print(len(ids))  # 檢查資料個數語法

```

```python
# 索引
ids = [1, 2, 3, 4, 5, 6]
print(ids[3])   
```

```python
# 修改索引位置的元素

ids = [1, 2, 3, 4, 5, 6]
ids[3] = 99
print(ids)


```

```python
# 刪除索引位置的元素
# list.pop(<index>)
ids = [1, 2, 3, 4, 5, 6]
ids.pop(3)
print(ids) 

```

```python
# 在指定索引 1， 新增元素 9 ,其餘元素往後移
# list.insert(<index>,<value>)
ids = [1, 2, 3, 4, 5, 6]
ids.insert(1, 9)
print(ids) 
```

```python
# 移除指定的值
# list .remove(<value>)
ids = [1, 2, 3, 4, 5, 5, 6]
ids.remove(5)
print(ids)
```

```python
ids = [1, 2, 3, 4, 5, 6] ### 重要
for _ in range(5):
    ids.insert(1, 9)
print(ids)

```

```python
while 9 in ids:
    ids.remove(9)
                             #為何可以不用寫 結束條件
print(ids)
```

```python
ids = [1, 2, 3, 4, 5, 5, 6]
for _ in range(5):
    ids.append(10)
print(ids)
# 計算 list 內的某個值 共有幾個
# list.count(<value>)
cnt_10 = ids.count(10)   # count不是改變list,所以要用一個變數ex:cnt_10來接收它
print(cnt_10)
```

```python
# 補充list的合併
ids1 = [1, 2, 3, 4]
ids2 = [5, 6, 7]

for i in ids2:    #用for迴圈也可以寫
    ids1.append(i)
print(ids1)
```

```python
# 補充list的合併 # method 
# list.extend()
ids1 = [1, 2, 3, 4]
ids2 = [5, 6, 7]
ids1.extend(ids2)
print(ids1)
print(ids2)
```

### list 搭配 for 迴圈

```python
# 建立一個 1~10 的 list
list1 = []
for i in range (1,11):
    list1.append(i)


print(list1)
```

```python
# 建立一個 1~50 的 list 只要偶數 
list2 = []
for i in range(1,51):
    if i % 2 ==0:
        list2.append(i)
print(list2)
```

```python
#建立一個 list 裡面要放迪增的list
#例如 [[1],[1, 2],[1, 2, 3]......[1, 2, 3..., 10]]
## 重要的觀念 *** # 

outside = []
inside = []
for i in range (1,11):
    inside.append(i)
    # print(inside)  
    outside.append(inside)

print(outside)
print(inside)
```

```python
outside = []
inside = []
for i in range (1,11):
    inside.append(i)
    # print(inside)  
    outside.append(inside.copy())  # 新增一個copy存放新的記憶體位置

print(outside)
print(inside)
```

```python
#建立一個 list 裡面要放迪增的list
#例如 [[1],[1, 2],[1, 2, 3]......[1, 2, 3..., 10]]
## 重要的觀念 *** # 

outside = []
for i in range (1,11):
    inside = []
    #print (f"i:{i}")
    for c in range(1, i+1):
        #print(f"c:{c}")
        inside.append(c)
    print(inside, id(inside)) # id()取得變數的記憶體位置
    outside.append(inside)
print(outside)
```

```python
# enumerate() 同時取得 index 與 value  ##補充補充 
# 在[] 搜尋value的index是多少 
ids = [5, 7, 9, 12, 5]
for index, value in enumerate(ids):
    print(f"index:{index},value:{value}")

```

### List comprehension 列表推導式
`new_list = [expression for item in iterable [if condition]]`  
  
`list的名稱 = [表達式 for 元素 in 迭代對象 [if 判斷式(...的條件)]]`

```python
#簡寫,執行效率比較高 , #實務上常用到 優化程式碼 #補充 
#建立一個1-10 的 list 
list1 = []
for i in range (1,11):
    list1.append(i)


print(list1)
```

```python
#list_comprehension 
num_list_1_to_10 = [i for i in range(1,11)]
print(num_list_1_to_10)
```

```python
# 1~10 的2次方的list
data = [i**2 for i in range(1,11)]
print(data)
```

```python
outside = []
for i in range (1,11):
    inside = []
    #print (f"i:{i}")
    for c in range(1, i+1):
        #print(f"c:{c}")
        inside.append(c)
    print(inside, id(inside)) # id()取得變數的記憶體位置
    outside.append(inside)
print(outside)
```

```python
outside = []
for i in range (1,11):
    inside = []
    inside = [c for c in range(1, i+1)] 
    outside.append(inside)
print(outside)
```

```python
outside = [] #進階
for i in range (1,11):
    outside.append([c for c in range(1, i+1)])
print(outside)
```

```python
#炫技  
outside = [[c for c in range(1, i+1)] for i in range(1,11)]
print(outside)
```

```python
# 建立一個 1~50 的 list 只要偶數  # 可以練習 
num_list = []
for i in range(1,51):
    if i % 2 ==0:
        num_list.append(i)
print(num_list)
```

```python
num_list = [i for i in range(1,51) if i% 2==0]
print(num_list)
```
