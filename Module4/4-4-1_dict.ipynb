```python
# 建立一個 value 含有 dict 的 dict
user_info = {
    "name": "Jarvis",
    "age": "20",
    "info": {
        "nickname": "Lu",
        "favorite_role": "ironman",
        "phone": [
            "0912345678",
            "0912345679"
        ]
    }
}
print(user_info)
```

```python
# 取得 nickname
print(user_info["info"])
print(user_info["info"]["nickname"])
```

```python
# 取得所有 phone_number
print(user_info["info"]["phone"])
for phone_number in user_info["info"]["phone"]:
    print(phone_number)
```

```python
'''
改寫 fruit_dict 每個水果需要 price, quantity/amount, color
apple 有 綠色 和 紅色
orange 有 橘色
tomato 有 紅色
'''
fruit_dict = {
        "apple": {
            "price" : 30,
            "quantity" : 100 ,
            "color" : ["green","red"]  
            },
        "orange": {
            "price" : 50,
            "quantity" : 30 ,
            "color" : ["orange"]   #建議用list , 與上面資料同屬性 
        },
        "tomato": {
            "price" : 50,
            "quantity" : 100 ,
            "color" : ["red"]
        } 
    }
print(fruit_dict)
```

##### dict 的合併

```python
# dict 的合併
# dict.update()
fruit_dict2 = {
    "kiwi" : {
        "price" : 30,
        "amount": 20,
        "color" : ["green"]
    },
}

fruit_dict.update(fruit_dict2)
print(fruit_dict)
```

```python
## 補充
print(fruit_dict)
fruit_dict3 = {
    **fruit_dict, # 把fruit_dict拆出來
    "kiwi" : {
        "price" : 30,
        "amount": 20,
        "color" : ["yellow"]
    },
}
print(fruit_dict3)
```

##### 另外的 print 方式 `pprint` (pretty-print) 

```python
# 從 Module 引入 function
from pprint import pprint
pprint(fruit_dict3) # 打印出來的資料沒有按照順序
```

```python
'''
新增顏色
apple 有新的品種 顏色要有黃色
tomato 有綠色
'''
fruit_dict3['apple']['color'].append('yellow')
fruit_dict3['tomato']['color'].append('green')

pprint(fruit_dict3)
```
