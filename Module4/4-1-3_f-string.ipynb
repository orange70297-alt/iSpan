```python
# 靠左對齊、靠右對齊

name = "Jarvis"
age = 18
msg = f"name:[{name:<10},age:[{age:>10}]]"
print(msg)
```

```python
# 靠左對齊、靠右對齊，字數不足，補「ㄏ」
# (沒寫補什麼，預設空格)
name = "Jarvis"
age = 18
msg = f"name:[{name:X<10},age:[{age:X>10}]]"
print(msg)

```

```python
# 置中對齊

name = "Jarvis"
age = 18
msg = f"name:[{name:^10},age:[{age:^6}]]"
print(msg)

```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)

name = "Jarvis"
age = 18
msg = f"name:[{name:.2}]"
print(msg)


```
