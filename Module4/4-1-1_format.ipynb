|語法|說明|
|:---:|:---:|
|%s |插入 文字 的值到字串中|
|%d |插入 整字  的值到字串中|
|%f |插入 浮.點數  的值到字串中|
| %%|輸出  (類似跳脫字元)|

```python
# 多組文字 ,相對不嚴謹 ,可以放數字
data1 = "Hello"
data2 = 100

msg = 'message: %s, %s !!!!'%(data1, data2)
print(msg) 
```

```python
# 整數 , 不可以為字串
age = 50
msg = 'i am %d years old.' %(age)
msg = '資料輸出的結果 Age= %d'%(age)

msg1 = 'My Age = %d' %(age)
print(msg)
print(age)
print(msg1)
```

```python
# 文字與整數
name= "Jarvis"
age=18
msg= " %s is %d years old" %(name, age)
print(msg)
```

```python
# 指定寬度 (維持 10 個字元長度，預設向右看齊) #證照考試考題
msg = "[%10s]" % "Hello" 
print(msg)

```

```python
# 靠左對齊 (維持 10 個字元長度，向左對齊)

msg = "[%-10s]" % "Hello" 
print(msg)
```

```python
# 指定浮點數位數

msg = '[%.2f]' % 3.14256
print(msg)

```

```python
msg = '[%.2f%%]' % 3.14256
print(msg)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)

msg = '[%.2s]' % "Hello"
print(msg)
```

```python
# 空白補 0

msg = '[%0.6f]' % 3.14
print(msg)
```
