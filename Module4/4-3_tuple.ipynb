```python
# tuple 初始化: 第一種
data = ("big","Data","is","Good")
print(data)
print(type(data))
```

```python
# tuple 初始化: 第二種   
data = "Jarvis" ,18 
print(data)
print(type(data))

name, age = "Jarvis" ,18     
print(name)
print(age)
print(type(age))
```

```python
data = "Jarvis" ,18 ,"AIPE02" #今天拿到tuple 如何取得裡面的資料
print(data)
print(type(data))

name,age,_class= data
print(name)
print(age)
print(_class)
print(type(_class))
```

```python
# 透過指定索引輸出值
data = "Jarvis" ,18 ,"AIPE02"
print(data)
print(data[0])
```

```python
# tuple 不可以修改指定索引的值
data[0] = 100
```

```python
# ( )可以修改指定索引的值

num_list = [1, 2, 3]
num_list[0] = 100
print(num_list)
```

```python
# 用 for 迴圈逐一輸出資料
data = "Jarvis" ,18,"AIPE02","11/2", "01:03"
for i in data:
    print(i, end=" ,")
```

```python
# 用 len 計算 tuple 資料個數
data = "Jarvis" ,18,"AIPE02"
# print(data)  # 實務上當資料量很大的時候會先以len確認資料筆數
print(len(data))
print(data[0])
```
