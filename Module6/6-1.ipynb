```python
# 類別名稱
class Car:
    # 類別屬性 (Class Attribute)
    wheel = 4
    # 建構函式 magic method
    def __init__(self, color, brand,wheels=4):
    # 實例屬性
        self.color = color
        self.brand = brand
        self.wheels = wheels
# 方法
    def info(self):
        print(f"color:{self.color},brand:{self.brand},wheels:{self.wheels}")

car1 = Car("red","toyota",wheels=6)
print(car1.color)
print(car1.brand)
print(car1.wheels)
car1.info()

print("="*50)
car2 = Car("black","benz")
print(car2.color)
print(car2.brand)
print(car2.wheels)

car2.info()
```

```python
class EditTxt:
    def __init__(self, file_path):
        self.file_path = file_path

    def read_file(self):
        with open(self.file_path, "r", encoding="utf-8") as f:
            doc = f.read()

        return doc 
    def insert_msg(self, msg):
        with open(self.file_path, 'a', encoding="utf-8")as f:
            f.write(msg)

txt_file = EditTxt("../Module2/2-1.py")
txt_str = txt_file.read_file()  # txt_str 只是一個變數
print(txt_str)

file2 = EditTxt("../Module9/9-1.py")
file2_str = file2.read_file()
file2.insert_msg("# 測試method 使用")
print(file2_str)        
```
