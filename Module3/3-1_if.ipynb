條件語句 if-elif-else

```python
'''
條件判斷
if ...
if ... else ...
if ... elif ... else ...

'''
```

##### if

```python
num = 20
if num > 10:
    print("num > 10" ) 
```

##### if...else...

```python
num = 2
if num > 10:
    print("num > 10" ) 
else:
    print("num 沒有大於 10")
```

##### if...elif... (else)

```python
#有先後重要的事情才會用 elif else來寫 
# else只發生在以上所有事情都沒有發生才執行, 也可比不用接else


num = 3
if num > 10:
    print("num > 10" ) 
elif num % 2 ==0:
    print("num 是偶數") 

elif type(num) == int:
    print("num 資料型態為 int")


else:
    print("num 沒有大於 10 且不為 int")
```
