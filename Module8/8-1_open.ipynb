##### 讀取檔案 read( )

```python
# 印出 50 個 "="
print("="*50)
"""
with 語句
通常用在 資料庫讀取,檔案讀取或寫入多線程
需要釋放資源的情況
"""

# 讀取檔案內容 - 常用的寫法 (執行完畢會自動關閉檔案)
with open(file="../Module2/2-1.py",mode="r",encoding="utf-8") as f : 
    doc = f.read()
    print(doc)

# 印出 50 個 "="

print("="*50)
```

```python
# 一行一行讀取

with open(file="../Module2/2-1.py",mode="r",encoding="utf-8") as f : 
    doc_list = f.readlines()
    print(doc_list)
```

```python
# 資料處理
# 去除所有\n 
with open(file="../Module2/2-1.py",mode="r",encoding="utf-8") as f : 
    doc_list = f.readlines()
    print(doc_list)

while "\n" in doc_list:
    doc_list.remove("\n")
print(doc_list)
new_list = []
for doc in doc_list:
    new_list.append(doc.strip("\n"))
print(doc_list)


                    
```

##### 寫入檔案 write( )

```python
# 文字內容
text = '''
傑出的人會努力規劃自己的人生，普通的人則是隨便命運怎麼安排   
一個是充實、有目的的過每一天，一個是隨興的過一天算一天
創業這條路 Michael E.Gerber
'''

# 寫入檔案
with open(file="test.txt",mode="w", encoding="utf-8") as f:
    f.write(text)


```

```python
# 寫入資料到最後一行 (附加文字在檔案內容最後一行)
msg = "People say that success just happens to you. It doesn't!"
with open(file="test.txt",mode="a", encoding="utf-8") as f:
    f.write(msg)

```
